```python
img_np = itk.array_view_from_image(img)
```

```python
ratio3 = round(int(meta['pixdim[3]'])/float(meta['pixdim[2]']))
ratio2 = round(float(meta['pixdim[1]'])/float(meta['pixdim[2]']))
```

```python
ratio3
```

```python
fig = plt.figure(figsize=(10,10))

ax = fig.add_subplot(1,3,1)
imgplot = plt.imshow(img_np[:,:,500])
ax = plt.gca()
ax.set_aspect(ratio3)
ax.set_title('Saggital view')


ax = fig.add_subplot(1, 3, 2)
imgplot = plt.imshow(img_np[10,:,:])
ax = plt.gca()
ax.set_aspect(ratio2)

ax.set_title('Transversal view')


ax = fig.add_subplot(1, 3, 3)
imgplot = plt.imshow(img_np[:,:,50])
ax = plt.gca()
ax.set_aspect(ratio3)
ax.set_title('Coronal view')
```
