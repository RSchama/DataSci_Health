# Heart Disorders that can be identified with ECGs

Electrocardiograms (or ECG or EKG) uses temporary electrodes on your chest and limbs to monitor, track and document your heart’s electrical activity (which controls your heartbeats) for diagnostic purposes. This is a quick, noninvasive test that doesn’t hurt. 

An EKG test is an easy way to get information to diagnose a problem with your heart. It doesn’t take long and doesn’t cause pain, but an electrocardiogram test can tell if you’ve had a heart attack, heart failure or heart damage. It can also tell if your heart rhythm isn’t normal or how well your pacemaker is working. The sinoatrial node starts the hearbeat with an electrical signal, the EKG reads that signal and tracks its impact on your heart as it contracts and relaxes with each heartbeat. We are intrested in looking at how much electrical activity there is, how strong it is and how much time passes between the different waves or peaks that represent the electrical impulses.   

1. Your upper heart chambers (atria), where your heartbeats start, create the first wave, or "P wave."
2. Your lower heart chambers (ventricles) create the next wave, called a QRS complex.
3. The third wave, or "T wave," shows your heart at rest or recovering after beating.

## Myocardial infarction

   This is the fancy name for a Heart Attack! It happens when one or more areas of the heart muscle don't get enough oxygen (), and this happens when blood flow to the heart is blocked somehow. Blood can be blocked by a blood clot, for example, which can happen when build up of plaques in the arteries breaks. If the blood and oxygen supply is cut off, muscle cells of the heart begin to die. Irreversible damage begins within 30 minutes of blockage. The heart muscle affected by the lack of oxygen no longer works as it should.  

But ischemia, or lack of blood going to the heart muscle, doesn't have to be accute, it can happen through a long period of time, even days. When it happens briafly the heart cells can recover, but when it lasts for a long period of time heart cells will start to die and that is what we call a myocardial infarction. It is critical to recognize ischemia in an early stage and in the first few hours after the myocardial infarction started, the ECG can be crucial. Nevertheless, detection of elevated serum cardiac enzymes is more important than ECG, for diagnistics.  

The ECG findings of an acute anterior myocardial infarction wall include: ST segment elevation in the anterior leads (V3 and V4) at the J point and sometimes in the septal or lateral leads, depending on the extent of the MI. This ST segment elevation is concave downward and frequently overwhelms the T wave

## Atrial fibrillation (AFib or AF)

   This is the most common type of treated heart arrhythmia, which is when the heart beats too slowly, too fast or in an irregular pattern. AFib happens when the normal beating in the two upper chambers of the heart (the two atria) is irregular, and blood doesn’t flow as well as it should from the atria to the lower chambers of the heart (the two ventricles). AFib may happen in brief episodes, or it may be a permanent condition. When this happens there is an increased risk of blood clots happening which increases a person’s risk for stroke, heart failure and other heart-related complicatios.  

   During atrial fibrillation, the heart's upper chambers (the atria) beat chaotically and irregularly — out of sync with the lower chambers (the ventricles) of the heart.

## Atrial flutter

   Atrial flutter is a type of heart rhythm disorder (arrhythmia) caused by problems in the heart's electrical system. In atrial flutter, the heart's upper chambers (atria) beat too quickly. This causes the heart to beat in a fast, but usually regular, rhythm. Atrial flutter is similar to AFib, but in the latter the heart beats in irregular patterns. People with atrial flutter have a heart rhythm that's more organized and less chaotic than that of AFib. Sometimes a person may have episodes of both types of arrhythmia. As AFib, this disorder can also increase the risk of stroke, heart failure and other complications.    


   A normal heartbeat begins with an electrical impulse from the **sinus node**, a small area in the heart's right upper chamber (right atrium). During Atrial flutter a short circuit allows the electrical impulse to quickly move around the right atrium, causing 240 to 340 contractions per minute. These rapids contractions prevent the chamber from filling up between beats. Blood left to pool in the upper chambers after an inefficient heartbeat increases the risk of clot formation. Also, If the heart beats very quickly over a long period of time, it may lead to cardiomyopathy, a weakening of the heart muscle. The ventricle also beats faster, but usually not as fast, there is a ratio of atrial beats to ventricular beats (for example 2:1). 

   Interestingly, when we feel our pulse we are feeling the left ventricular contractions (as it pumps blood to your arteries), and a normal heart rate is between 60 to 100 beats per minute (BPM).

```python
# plot function
def plotECG(center, width, c_file):
    start = center - width
    stop = center + width
    
    fig, ax = plt.subplots()
    a = ECGRecord.from_wfdb(path + c_file)
    plt.title('ECGRecord', fontsize=10)
    plt.xlabel('time')
    plt.ylabel('signal')
    plt.plot(range(start,stop),a.p_signal[0,start:stop], '--', label='lead1')
    plt.plot(range(start,stop),a.p_signal[1,start:stop],'--', label='lead2')
    plt.plot(range(start,stop),a.p_signal[0,start:stop]+a.p_signal[1,start:stop], label='lead1 + lead2')
    plt.legend(loc='upper right')
    plt.show()
```
