```python
import itk
import numpy as np
import matplotlib.pyplot as plt

import ipywidgets as widgets
from ipywidgets import interact, interactive, fixed, interact_manual
```

```python
imgPath = r"TestData/"
```

```python
img2 = itk.imread(imgPath + "Prostate.mha")
```

```python
img_np2 = np.asarray(img2)
print(img2.shape)
```

```python
fig = plt.figure(figsize=(10,20))

ax = fig.add_subplot(1,3,1)
imgplot = plt.imshow(img_np2[:,:,60])
ax = plt.gca()
ax.set_aspect(40)
ax.set_title('Saggital view')


ax = fig.add_subplot(1, 3, 2)
imgplot = plt.imshow(img_np2[10,:,:])
ax.set_title('Transversal view')


ax = fig.add_subplot(1, 3, 3)
imgplot = plt.imshow(img_np2[:,800,:])
ax = plt.gca()
ax.set_aspect(40)
ax.set_title('Coronal view')
```

```python
def MRIPlotting(trans, coronal, sag):
    fig = plt.figure(figsize=(10,20))

    ax = fig.add_subplot(1,3,1)
    imgplot = plt.imshow(img_np2[:,coronal,:])
    ax = plt.gca()
    ax.set_aspect(40)
    ax.set_title('Coronal view')


    ax = fig.add_subplot(1, 3, 2)
    imgplot = plt.imshow(img_np2[trans,:,:])
    ax.set_title('Transversal view')


    ax = fig.add_subplot(1, 3, 3)
    imgplot = plt.imshow(img_np2[:,:,sag])
    ax = plt.gca()
    ax.set_aspect(40)
    ax.set_title('Saggital view')
    
```

```python
trans_slider = widgets.IntSlider(min=0,max=27,step=1,value=10)
coronal_slider = widgets.IntSlider(min=0,max=1024,step=10,value=500)
sag_slider = widgets.IntSlider(min=0,max=1024,step=10,value=50)

widgets.interact(MRIPlotting,trans=trans_slider, coronal=coronal_slider, sag=sag_slider)
```
